# the most winning team per season

```python
winner=matches.groupby(['season'])['winner'].value_counts()
winner.head(10)
```

```python
matches.groupby(['season'])['winner'].value_counts().sort_values(ascending=True).tail(10).plot(kind = 'bar', figsize = (20,8),color=['r','g','b','c','k','y','m'])
plt.title("Winners of IPL across 11 seasons",fontsize=18,fontweight="bold")
plt.xlabel("Teams", size = 25)
plt.ylabel("Winning Count", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

# Number of teams play every year

```python
matches.groupby('season')['team1'].nunique().plot(kind = 'bar', figsize=(15,5),color='m')
plt.title("Number of teams participated each season ",fontsize=18,fontweight="bold")
plt.ylabel("Count of teams", size = 25)
plt.xlabel("Season", size = 25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

# Number of Matches played every season

```python
#plot.figure(figsize=(15,5))
#sns.countplot('season',data=matches)
plt.figure(figsize=(15,5))
sns.countplot('season', data = matches)
plt.title("number of matches play every season",fontsize=20,fontweight='bold')
plt.xlabel("Count",size=15)
plt.ylabel("Season",size=15)
plt.xticks(size=20)
plt.yticks(size=20)
```

# Conclusion: Number of Team between 2011 and 2013 is more than other season so matches in this seasons more than others

# Locations which has hosted most number of IPL matches .

```python
matches.venue.value_counts().sort_values(ascending = True).tail(10).plot(kind ='barh',figsize=(20,8),fontsize=20,color=['r','g','b','c','k','y','m'])
plt.title("Locations which has hosted most number of IPL matches",fontsize=25,fontweight="bold")
plt.ylabel("Location", size = 25,fontweight='bold')
plt.xlabel("Frequency", size = 25,fontweight='bold')
```
